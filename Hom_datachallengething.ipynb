```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline


df = pd.read_csv('sorted data.csv')
race = df.sort_values(by = ['year'])
df2018=df.where(df['year']==2018).dropna().drop(columns='year')
df2019=df.where(df['year']==2019).dropna().drop(columns='year')
df2021=df.where(df['year']==2021).dropna().drop(columns='year')
df2022=df.where(df['year']==2022).dropna().drop(columns='year')

racedf2018 = df2018.sort_values(by = ['Race'])
racedf2019 = df2019.sort_values(by = ['Race'])
racedf2021 = df2021.sort_values(by = ['Race'])
racedf2022 = df2022.sort_values(by = ['Race'])

data2018 = (
    racedf2018
    .groupby(['Race'])
    .agg(
        will30_app=('will30_app', 'mean'),
        willever_tixtalk=('willever_tixtalk', 'mean'),
        willever_tixtext=('willever_tixtext', 'mean'),
        willever_crashtalk=('willever_crashtalk', 'mean'),
        willever_crashtext=('willever_crashtext', 'mean')
    )
    .reset_index()
)
data2019 = (
    racedf2019
    .groupby(['Race'])
    .agg(
        will30_app=('will30_app', 'mean'),
        willever_tixtalk=('willever_tixtalk', 'mean'),
        willever_tixtext=('willever_tixtext', 'mean'),
        willever_crashtalk=('willever_crashtalk', 'mean'),
        willever_crashtext=('willever_crashtext', 'mean')
    )
    .reset_index()
)
data2021 = (
    racedf2021
    .groupby(['Race'])
    .agg(
        will30_app=('will30_app', 'mean'),
        willever_tixtalk=('willever_tixtalk', 'mean'),
        willever_tixtext=('willever_tixtext', 'mean'),
        willever_crashtalk=('willever_crashtalk', 'mean'),
        willever_crashtext=('willever_crashtext', 'mean')
    )
    .reset_index()
)
data2022 = (
    racedf2022
    .groupby(['Race'])
    .agg(
        will30_app=('will30_app', 'mean'),
        willever_tixtalk=('willever_tixtalk', 'mean'),
        willever_tixtext=('willever_tixtext', 'mean'),
        willever_crashtalk=('willever_crashtalk', 'mean'),
        willever_crashtext=('willever_crashtext', 'mean')
    )
    .reset_index()
)

heatmap2018 = data2018.set_index('Race')
heatmap2019 = data2019.set_index('Race')
heatmap2021 = data2021.set_index('Race')
heatmap2022 = data2022.set_index('Race')
```
